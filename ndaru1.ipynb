```python
#@title Visualisasi - Part 2: Matplotlib

import pandas as pd
import matplotlib.pyplot as plt

tabel = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/produktivitas_karyawan.csv", index_col="Bulan") / 1000

fig, ax = plt.subplots(1, 1, figsize=(12, 6))
fig.suptitle("Best Employee 2019", x=0.028, y=1.05, fontsize=24, fontweight='bold', ha="left")
ax.set_title("Lima karyawan dengan produktivitas tertinggi\n", fontsize=18, ha="right")

colors = ["tab:gray", "tab:gray", "tab:gray", "tab:gray", "darkblue"]
line_widths = [ 1.5, 2, 1, 2.5, 3.]
lp = tabel.plot(kind="line", ax=ax, xlim=[0, 11], ylim=[10, 40], color=colors,
xticks=range(tabel.shape[0]), yticks=[10, 20, 30, 40])

add_lbl_pos = [ 0.3, 0.5, 1.2, -1.0, 0.5]
add_avg_pos = [-0.7, -0.5, 0.2, -2.0, -0.5]
for i, item in enumerate(lp.get_lines()):
	item.set_linewidth(line_widths[i])
	lbl = item.get_label()
	ax.annotate(lbl, (11, tabel.loc["Desember", lbl]), (11.2, tabel.loc["Desember", lbl] + add_lbl_pos[i]),
fontweight="bold", fontsize=12, color=colors[i], va="center",
arrowprops={"arrowstyle": "-", "color": colors[i]})
ax.annotate("(avg: %.4f)" % tabel[lbl].mean(), (11, tabel.loc["Desember", lbl]),
(11.2, tabel.loc["Desember", lbl] + add_avg_pos[i]),
fontsize=11, color=colors[i], va="center")
highest_prod = tabel["Karyawan E"].max()
ax.annotate("Produktivitas\ntertinggi\n%.4f" % highest_prod, (8, highest_prod), (7.75, highest_prod+1.5),
ha="left", color=colors[-1], arrowprops={"arrowstyle": "-", "color": colors[-1]})
ax.set_xticklabels([item[:3] for item in tabel.index], fontsize=12)
ax.set_yticklabels([str(i) for i in [10, 20, 30, 40]], fontsize=12)
ax.set_xlabel("")
ax.set_ylabel("Jumlah produk (ribuan)\n", fontsize=14)
ax.spines["top"].set_color("none")
ax.spines["right"].set_color("none")
ax.spines["bottom"].set_color("none")
ax.get_legend().remove()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import io
# Simulating CSV data loading (since data is provided directly)
csv_data = """Kelompok Usia,Laki-laki,Perempuan
51-55 tahun,417,199
46-50 tahun,502,319
41-45 tahun,516,420
36-40 tahun,353,332
31-35 tahun,198,208
26-30 tahun,124,156
21-25 tahun,45,53
"""

# Load CSV data into DataFrame
df = pd.read_csv(io.StringIO(csv_data))

# Calculate total employees per age group
df['Total'] = df['Laki-laki'] + df['Perempuan']

# Identify age groups with even totals
df['Is_Even'] = df['Total'] % 2 == 0

# Print analysis
print("Analisis Kelompok Usia dengan Jumlah Total Karyawan Genap:")
even_groups = df[df['Is_Even']]
for index, row in even_groups.iterrows():
    print(f"Kelompok Usia {row['Kelompok Usia']}: Total = {row['Total']} (Genap)")

# Visualization: Stacked bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(df))

# Plot bars for male and female
plt.bar(index, df['Laki-laki'], bar_width, label='Laki-laki', color='skyblue')
plt.bar(index, df['Perempuan'], bar_width, bottom=df['Laki-laki'], label='Perempuan', color='lightcoral')

# Highlight age groups with even totals
for i, row in df.iterrows():
    if row['Is_Even']:
        plt.text(i, row['Total'] + 10, f'Total: {row["Total"]} (Even)', ha='center', fontsize=10, color='green')

# Customize plot
plt.xlabel('Kelompok Usia')
plt.ylabel('Jumlah Karyawan')
plt.title('Distribusi Karyawan Berdasarkan Kelompok Usia dan Jenis Kelamin')
plt.xticks(index, df['Kelompok Usia'], rotation=45)
plt.legend()
plt.tight_layout()

# Save the plot
plt.savefig('employee_distribution.png')
print("\nVisualisasi telah disimpan sebagai 'employee_distribution.png'")
```
